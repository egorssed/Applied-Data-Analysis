```python
def process_script(script):
    episodes=script.split('>>')
    scenes_split=[episode.split('>') for episode in episodes]
    
    script_rows=list()
    
    for episode in tqdm(scenes_split):
        episode_rows=process_episode(episode)
        
        if episode_rows==[]:
            continue
        
        script_rows=script_rows+episode_rows
        
    return script_rows
    
def process_episode(episode):
    #Obtain number of Series and number of Episode
    title_numbers=[int(s) for s in episode[0].split() if s.isdigit()]
    if len(title_numbers)<2:
        #return empty df and go further
        return []
    
    [Series,Episode]=title_numbers[:2]
    
    #scenes[:] - all the scenes
    scenes=[scene.split('\n') for scene in episode[1:]]
    
    Complete_episode_rows=list()
    
    for scene_number,scene in enumerate(scenes):
        #Get list of [location,character,line]
        scene_rows=process_scene(scene)
        
        Numbers=[Series,Episode,scene_number]
        
        Complete_scene_rows=[Numbers+speech for speech in scene_rows]
        
        #Append data to episode row
        Complete_episode_rows=Complete_episode_rows+Complete_scene_rows
    
    return Complete_episode_rows

def process_scene(scene):
    speeches=list()
    
    #scenes[:][0] - locations
    location=scene[0]
        
        
    #scenes[i][j>0] - lines, i.e. 'Leonard: Okay.'
    #0-location
    #-1 always ' '
    for speech in scene[1:]:
        charact_sep=speech.find(':')
            
        character=speech[:charact_sep]
            
        #+2 to avoid : and space 
        line=speech[charact_sep+1:]
            
        speeches.append([location,character,line])
    
    return speeches
```

```python
def confusion_matrix(y_pred,y_true):
    
    TP=((y_pred==1) & (y_true==1)).sum()
    FN=((y_pred==0) & (y_true==1)).sum()
    
    TN=((y_pred==0) & (y_true==0)).sum()
    FP=((y_pred==1) & (y_true==0)).sum()
    
    confusion=np.array([[TP,FP],[FN,TN]])
    confusion=pd.DataFrame(confusion)
    confusion.index=pd.MultiIndex.from_tuples([('Predicted','True'),('Predicted','False')])
    confusion.columns=pd.MultiIndex.from_tuples([('Actual','True'),('Actual','False')])
    return confusion

def characteristics(confusion,print_values=False):
    [[TP,FP],[FN,TN]]=confusion.to_numpy()

    if print_values:
        print('Accuracy {:.2f}'.format((TP+TN)/confusion.sum().sum()))
        print('Precision positive {:.2f}, Recall positive {:.2f}, F1 {:.2f}'.format(
            TP/(TP+FP),TP/(TP+FN),TP/(TP+0.5*(FP+FN))))
        print('Precision negative {:.2f}, Recall negative {:.2f}, F1 {:.2f}'.format(
            TN/(TN+FN),TN/(TN+FP),TN/(TN+0.5*(FP+FN))))
        
    return [(TP+TN)/confusion.sum().sum(),(TP/(TP+FP),TP/(TP+FN),TP/(TP+0.5*(FP+FN))),(TN/(TN+FN),TN/(TN+FP),TN/(TN+0.5*(FP+FN)))]
```

```python
confusion=confusion_matrix(test_predictions,df_Test.Sheldon)
confusion
```

```python
_=characteristics(confusion,print_values=True)
```
